#### Derive CDW from OD, where missing

```python
sns.scatterplot(
    data=offline_df,
    y="CDW_avg [g/L]",
    x="OD"
)

xy_data = offline_df[["CDW_avg [g/L]", "OD"]]
xy_data.dropna(inplace=True)

x = xy_data["OD"]
y = xy_data["CDW_avg [g/L]"]

#regression part
slope, intercept, r_value, p_value, std_err = stats.linregress(x,y)

line = slope*x+intercept
plt.plot(x, line, 'r', label='y={:.2f}x+{:.2f}'.format(slope,intercept))

plt.legend(fontsize=9)
```

```python
CDW_list = []
correlated_CDW_list = []

for i in range(0, len(offline_df)):
    if (math.isnan(offline_df.iloc[i]["CDW_avg [g/L]"])) & (not math.isnan(offline_df.iloc[i]["OD"])):
        CDW_list.append(0.29*offline_df.iloc[i]["OD"])
        correlated_CDW_list.append("yes")
        
    elif (math.isnan(offline_df.iloc[i]["CDW_avg [g/L]"])) & (math.isnan(offline_df.iloc[i]["OD"])):
        CDW_list.append(0.29*offline_df.iloc[i]["OD"])
        correlated_CDW_list.append("no_biomass_data")
        
    else:
        CDW_list.append(offline_df.iloc[i]["CDW_avg [g/L]"])
        correlated_CDW_list.append("no")
        
offline_df["correlated_CDW"] = correlated_CDW_list
offline_df["CDW_avg [g/L]"] = CDW_list
```

### Advanced calculations

#### Functions

```python
def lost_compount_calc(new_lost_col,new_lost_cumsum_col,measurement_col):
    offline_df[new_lost_col] = offline_df[measurement_col] * offline_df['Sample volume (mL)']/1000
    offline_df[new_lost_cumsum_col] = offline_df.groupby(['Parent Culture']).cumsum()[new_lost_col]
    offline_df[new_lost_cumsum_col] = offline_df.groupby('Parent Culture').shift()[new_lost_cumsum_col].fillna(0)
    
def net_formation(lost_cumsum_col, residual_col, new_net_col, feed_conc, init_amount):
    # Formed amount = residual_conc*volume + lost_amount - initial_amount - feed_conc*feed_vol
    offline_df[new_net_col] = offline_df[residual_col]*offline_df['Liquid_volume']/1000 + offline_df[lost_cumsum_col] - init_amount - feed_conc*offline_df['Feed_volume']/1000
    
def net_consumption(lost_cumsum_col, residual_col, new_net_col, feed_conc, init_amount):
    # Formed amount = residual_conc*volume + lost_amount - initial_amount - feed_conc*feed_vol
    offline_df[new_net_col] = -offline_df[residual_col]*offline_df['Liquid_volume']/1000 - offline_df[lost_cumsum_col] + init_amount + feed_conc*offline_df['Feed_volume']/1000
```
